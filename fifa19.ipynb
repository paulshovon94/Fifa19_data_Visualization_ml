```python
# comparison bitween sprintSpeed and shotpower among top 10 players in Line Plot?
fifa.loc[:10].SprintSpeed.plot(kind='line',color='blue',label='SprintSpeed',grid='True',linewidth=2,alpha =.5,linestyle=':',figsize = (10,8))
fifa.loc[:10].ShotPower.plot(kind='line',color='red',label='ShotPower',grid='True',linewidth=2,alpha = .5,linestyle='-.',figsize = (10,8))
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```
